```python
import glob
import os
import pandas as pd
from scipy import stats
import numpy as np
```

```python
# Identifying directory of lab sheets for calculations

path = str(glob.os.getcwd())

user = path.split('\\')[2]

folder = "Lab_Sheets/"

datasheets = '/Users/' + user + '/Desktop/Coding/StandardLab/' + folder

glob.os.chdir(datasheets)



# Setting loop for iterating through files WITHIN folders IN the specified folder

for x, folder, file in os.walk(datasheets):
    
    for loc in file:
        
        if loc.split('_')[2] == "NIT":
            print('Yes ' + str(loc))
            
        elif loc.split('_')[2] == "URE":
            print("yepp " + str(loc))
            
        else:
            continue
```

```python
for x, folder, file in os.walk(datasheets):
    
    for loc in file:
        
        if loc.split('_')[2] == "NIT":
            
            DF = pd.read_csv(str(x) + '/' +str(loc))
            #print(DF)
    
            nit_df = DF[['Sample_ID', 'Absorbance']]

            nit_g = nit_df.groupby('Sample_ID')

            nit_final=nit_df.merge(nit_g.mean(),on='Sample_ID')

            nit_final.rename(columns = {'Absorbance_x':'Each', 'Absorbance_y':'Mean'}, inplace=True)
            
            nit_final['Type'] = DF['Type']
                 
            c=[]
            m=[]
            
            for i, row in nit_final.iterrows():
                
                if row['Type'] == 'C':
            
                    c.append(float(row['Sample_ID']))
                    m.append(float(row['Mean']))
                    
                    
            slope, intercept, r_value, p_value, std_err = stats.linregress(m, c)
            
            conc = []
            
            for i, row in nit_final.iterrows():
                
                conc.append((float(row['Mean'])*slope + intercept))
            
            nit_final['Sample_Conc'] = conc
            
 #         elif loc.split('_')[2] == "URE":
 #             print("yepp " + str(loc))
 #         else:
 #             continue
            print(nit_final)
    
```
